```python
import math

n = 2500

f1 = n**(math.sqrt(math.log(n)))
f2 = 7*n - 2
f3 = 100*math.sqrt(n) + 2*n**(1/4) - 100
f4 = 2**((math.log(n))**2)
f5 = 2**(2*math.log(n)) + n**(1.1)
f6 = 3**n
f7 = n*math.log(n)

# Store the functions in a dictionary
functions = {
    "f1": f1,
    "f2": f2,
    "f3": f3,
    "f4": f4,
    "f5": f5,
    "f6": f6,
    "f7": f7
}

# Sort the functions by their values
sorted_functions = sorted(functions.items(), key=lambda x: x[1])

# Print the sorted functions in the desired format
print(" << ".join([f"{name}" for name, value in sorted_functions]))

# Print some space
print("\n")

# Print the results
print(f"f1 = {f1}")
print(f"f2 = {f2}")
print(f"f3 = {f3}")
print(f"f4 = {f4}")
print(f"f5 = {f5}")
print(f"f6 = {f6}")
print(f"f7 = {f7}")
```

```python
import math

n = 2500

f1 = 2**n
f2 = 3**(n//2)

# Store the functions in a dictionary
functions = {
    "f1": f1,
    "f2": f2
}

# Sort the functions by their values
sorted_functions = sorted(functions.items(), key=lambda x: x[1])

# Print the sorted functions in the desired format
print(" << ".join([f"{name}" for name, value in sorted_functions]))

# Print some space
print("\n")

# Print the results
print(f"f1 = {f1}")
print(f"f2 = {f2}")
```
